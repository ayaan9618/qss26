```python
# 4-qubit GHZ state on IBM Quantum (least busy backend)
from qiskit import QuantumCircuit
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_ibm_runtime import QiskitRuntimeService, SamplerV2 as Sampler

# Build GHZ circuit
qc = QuantumCircuit(4)
qc.h(0)
qc.cx(0, 1)
qc.cx(1, 2)
qc.cx(2, 3)
qc.measure_all()

# Connect to IBM Quantum and find least busy backend
service = QiskitRuntimeService()
backend = service.least_busy(min_num_qubits=4, operational=True)
print(f"Submitting to: {backend.name}")

# Transpile to the backend's native gate set
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
qc_t = pm.run(qc)

# Run the circuit
sampler = Sampler(mode=backend)
job = sampler.run([qc_t], shots=1024)
print(f"Job ID: {job.job_id()}")
print("Job submitted! Check the IBM Cloud tab in the Jobs panel.")
```
